### Linear regression, Lasso regression, Ridge regression, Elasticnet regression-

#### We take FWI (Fire Weather Index) as the dependent feature and predict it using other independent features. As FWI is a continuous variable, so it becomes a regression problem.

```python
# we won't need day, month, year variables so we drop them
df= df.drop(["day", "month", "year"], axis=1)
```

```python
df.head()
```

```python
X= df.drop("FWI", axis=1)              # all the independent variables except FWI are taken in X
y= df["FWI"]                           # y contains the dependent varaible FWI
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
# TRAIN-TEST SPLIT-
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size= 0.25,random_state= 42)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
# FEATURE SELECTION BASED ON CORRELATION-
X_train.corr()
```

```python
# CHECK FOR MULTICOLLINEARITY-
# Multicollinearity- when independent features are highly correlated with each other. So, they act similar and we can drop one of them.
plt.figure(figsize=(12,10))
sns.heatmap(X_train.corr(), annot=True)
```

```python
X_train.corr().columns
```

```python
def correlation(dataset, threshold):
    col= set()
    corr_matrix= dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if corr_matrix.iloc[i,j]>threshold:
                col.add(corr_matrix.columns[i])
    return col
    
```

```python
features= correlation(X_train,0.85)
features
```

```python
# Drop these features where correlation is greater than 0.85-
X_train.drop(features, axis=1, inplace=True)
X_test.drop(features, axis=1, inplace=True)
```

```python
X_train.shape, X_test.shape
```

```python
#FEATURE SCALING OR STANDARDIZATION-
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
# Box plots to understand the effect of Standard Scaler-
plt.subplots(figsize=(15,5))
plt.subplot(121)
sns.boxplot(data=X_train)
plt.title("X_train before scaling")
plt.subplot(122)
sns.boxplot(data=X_train_scaled)
plt.title("X_train after scaling")
```

#### Linear Regression Model-

```python
from sklearn.linear_model import LinearRegression
linreg= LinearRegression()
linreg.fit(X_train_scaled, y_train)
y_pred= linreg.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
mae= mean_absolute_error(y_test, y_pred)
score= r2_score(y_test, y_pred)
print("Mean absolute error is:", mae)
print("R2 score is:", score)
plt.scatter(y_test, y_pred)
```

```python
# The accuracy given by R2_score is very high i.e. nearly 98.5%
# the scatter plot looks linear too suggesting that the actual values (y_test) and the predicted values (y_pred) are nearly equal.
```

#### Lasso Regresion model-

```python
from sklearn.linear_model import Lasso
lasso= Lasso()
lasso.fit(X_train_scaled, y_train)
y_pred= lasso.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
mae= mean_absolute_error(y_test, y_pred)
score= r2_score(y_test, y_pred)
print("Mean absolute error is:", mae)
print("R2 score is:", score)
plt.scatter(y_test, y_pred)
```

#### Cross Validation Lasso-

```python
# validation set- a subset is taken from the training set called validation set which is used for hyperparameter tuning
```

```python
from sklearn.linear_model import LassoCV
lassocv= LassoCV(cv=5)        # cv= 5 means cross validate 5 times and validation set= 182/5= 36 data points
lassocv.fit(X_train_scaled, y_train)
y_pred= lassocv.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
mae= mean_absolute_error(y_test, y_pred)
score= r2_score(y_test, y_pred)
print("Mean absolute error is:", mae)
print("R2 score is:", score)
plt.scatter(y_test, y_pred)
```

```python
# we can see that the accuracy given by r2_score was 94.9% in lasso regression model but in lassocv model the accuracy shooted up to 98.2%
# so by cross validation, the accuracy increases.
```
